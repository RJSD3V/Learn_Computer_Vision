##Quicky Tutorial
Using the MNIST_Sample Data, here is a quicky model in a few lines of code. 

```
from fastai.vision import *

```

```
import pathlib
path='/content/cifar10.tgz'
```

```
import tarfile
tar=tarfile.open(path)
tar.extractall('/content/sample_data/')
```

```
from pathlib import *
cifar_path=PosixPath('/content/sample_data/cifar10');cifar_path
```

```
cifar_train=cifar_path/'train'
cifar_test=cifar_path/'test'
```

```
cifar_labels=['airplane',
 'automobile',
 'bird',
 'cat',
 'deer',
 'dog',
 'frog',
 'horse',
 'ship',
 'truck',
 '']
```

```
data=ImageDataBunch.from_folder(cifar_path,train='train',valid='test',classes=cifar_labels,test=None,ds_tfms=get_transforms(),size=224,bs=64)
```

```
data.show_batch(5,figsize=(7,6))
```

```
print(data.classes)
print(data.c)
```

```

learn=cnn_learner(data,models.resnet34,metrics=accuracy)
```

```
learn.fit(5)
```
